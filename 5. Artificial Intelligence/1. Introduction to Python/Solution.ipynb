#### Question 3

```python
print("Enter your age:")
age = input()

print("Are you a student?")
student = input()

if int(age) >= 18 and int(age) <= 60:
    if student.lower() == "yes":
        print("Valid! Age between 18 and 60 and a student.")
    else:
        print("Invalid! Not a student")
else:
    if student.lower() == "yes":
        print("Invalid! Age not between 18 and 60.")
    else:
        print("Invalid! Neither age betweeen 18 and 60 nor a student.")
```

#### Question 4

```python
print("Enter your Marks:")
num = input()
marks = int(num)

if marks >= 90:
    print("Grade: A")
elif marks >= 80:
    print("Grade: B")
elif marks >= 70:
    print("Grade: C")
elif marks >= 60:
    print("Grade: D")
else:
    print("Grade: F")
```
