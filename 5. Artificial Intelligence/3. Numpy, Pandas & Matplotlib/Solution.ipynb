### Question 4

```python
df3 = pd.read_csv('Bank Dataset.csv')
df3.head()      
```

```python
print("Minimum Account Balance: {}" .format(df3['AccountBalance'].min()))
print("Maximum Account Balance: {}" .format(df3['AccountBalance'].max()))
print("Sum of all Account Balances: {}" .format(df3['AccountBalance'].sum()))
print("Cumulative Sum of Account Balances:")
for val in df3['AccountBalance'].cumsum():
    print(" - {}" .format(val))
```

```python
print("Null values in each column:")
for col in df3.columns:
    print(" - {}: {}" .format(col,df3[col].isna().sum()))
```

##### The missing values tell the completeness of data, and missing values need to be filled before performing further processing
##### The min, max, sum and cumsum values tell the range of the data and how is it spread 


### Question 5

```python
from sklearn.datasets import load_iris
iris = load_iris(as_frame=True)
```

```python
df4 = pd.DataFrame(data=iris.data, columns=iris.feature_names)

print("First 10 rows of the Iris dataset:")
df4.head(10)
```

```python
df4['Species(Target)'] = iris.target.map(lambda x: iris.target_names[x])
df4.head(10)
```

```python
print("Target Column: {}" .format(df4.columns[-1]))

specie_names = df4['Species(Target)'].unique()
print("Values: {} \n" .format(specie_names))


print("Feature Columns:")
for i in (list(df4.columns)[:-1]):
    print(" - {}" .format(i))
```

```python
print("Mean Petal Length by Specie: \n")

mean_by_specie = df4.groupby('Species(Target)').mean()
mean_by_specie['petal length (cm)']
```

```python
Salaries = [2500, 3000, 4000, 3500, 4500, 5000, 3200]

plt.boxplot(Salaries)
plt.grid()
plt.show()
```

```python
series = pd.Series(Salaries)
print("Minimum: {}" .format(series.min()))
print("Q1 (25%): {}" .format(series.quantile(0.25)))
print("Median (50%): {}" .format(series.median()))
print("Q1 (75%): {}" .format(series.quantile(0.75)))
print("Maximum: {}" .format(series.max()))
```
