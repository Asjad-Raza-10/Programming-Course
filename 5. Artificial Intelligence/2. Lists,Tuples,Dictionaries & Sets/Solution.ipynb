#### Question 1

```python
fruits = ['apple', 'banana', 'cherry', 'date', 'fig']

print("First element: {}".format(fruits[0]))
print("Last element: {}".format(fruits[-1]))
print("First three elements: {}".format(fruits[:3]))
print("Last two elements: {}".format(fruits[-2:]))
```

#### Question 2

```python
fruits = ['apple', 'banana', 'cherry', 'date', 'fig']

fruits[1] = 'blueberry'
del fruits[-1]
fruits.append('grape')

print("Modified list: {}".format(fruits))
```

#### Question 3

```python
numbers = [10, 5, 7, 3, 9]

print("Sum of the list: {}".format(sum(numbers)))
print("Maximum number: {}".format(max(numbers)))
print("Minimum number: {}".format(min(numbers)))

numbers.sort()
print("Sorted list: {}".format(numbers))
```

#### Question 4

```python
squares_of_even = [n**2 for n in range(1, 11) if n % 2 == 0]
print("Squares of even numbers: {}".format(squares_of_even))

fruits = ['apple', 'blueberry', 'cherry', 'date', 'grape']
short_fruits = [fruit for fruit in fruits if len(fruit) < 6]
print("Short fruits: {}".format(short_fruits))
```

#### Question 7

```python
student = {'name': 'Ali', 'age': 20, 'grade': 'A'}

print("Name: {}".format(student['name']))

student['city'] = 'Lahore'
student['grade'] = 'A+'
del student['age']

print("Modified dictionary: {}".format(student))
```

#### Question 8

```python
scores = {'Math': 90, 'Physics': 85, 'Chemistry': 92}

print("Keys: {}".format(list(scores.keys())))
print("Values: {}".format(list(scores.values())))
print("Total: {}".format(sum(scores.values())))

hash = {'word': 'Math', 'count': scores['Math']}
print("Ali scored %(count)d in %(word)s" % hash)
```

#### Question 9

```python
tools = {'Photoshop', 'Illustrator', 'Acrobat'}

tools.add('Premiere')
tools.remove('Acrobat')

print("Trying tools.add(['InDesign', 'Bridge']) gives Error: unhashable type: 'list'")

tools.clear()
print("Tools: {}".format(tools))
```

#### Question 10

```python
DS_skills = {'Python', 'R', 'SQL', 'Git'}
DE_skills = {'Python', 'Java', 'SQL', 'Hadoop'}

print("Union: {}".format(DS_skills | DE_skills))
print("Intersection: {}".format(DS_skills & DE_skills))
print("Difference: {}".format(DS_skills - DE_skills))
print("Symmetric Difference: {}".format(DS_skills ^ DE_skills))
```
